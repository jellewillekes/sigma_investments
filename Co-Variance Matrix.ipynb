```python
import math
import numpy as np
import matplotlib
import pandas_datareader as dr
import pandas as pd
import datetime as dt
from datetime import datetime
```

```python
start = '2018-01-01'
today = datetime.today().strftime('%Y-%m-%d')
tickers = pd.read_excel ('SigmaInput.xlsx')["Ticker"]

price_data = pd.DataFrame()

for t in tickers:
    price_data[t] = dr.data.get_data_yahoo(t, start =start, end =today)['Adj Close']

return_data = (price_data/price_data.shift(1))-1
price_data
```

```python
cov_matrix_annual = return_data.cov()*252
cov_matrix_annual.to_excel('Co-VarMatrix.xlsx')
print(cov_matrix_annual)
```

```python
correlation_matrix = return_data.corr()
correlation_matrix.to_excel('CorrelationMatrix.xlsx')
correlation_matrix
```
